```python
import numpy as np
from matplotlib import pyplot as plt

x=np.linspace(-10,10,10000)
y=x**2-4*x+4
plt.plot(x,y)
plt.show()
```

```python

y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,color='red')

plt.plot(x,z,color='blue')
plt.show()
```

```python
plt.subplot(2,2,1)
y=x**3
plt.plot(x,y)
plt.subplot(2,2,2)
y=np.sin(x)
plt.plot(x,y)
plt.subplot(2,2,3)
y=np.e**x
plt.plot(x,y)
plt.subplot(2,2,4)

y=np.where(x<0,0,np.log(x))
plt.plot(x,y)
```

```python
x=np.random.uniform(0,10,100)
y=np.random.uniform(0,10,100)
plt.scatter(x,y)
plt.show()
```

```python
x=np.random.normal(0,1,1000)
plt.hist(x,bins=30,alpha=0.5)
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
```

```python
x=np.linspace(-5,5,1000)
y=np.linspace(-5,5,1000)
def f(x,y):
    return np.cos(x**2+y**2)
plt.figure()
ax=plt.axes(projection='3d')
ax.plot3D(x,y,f(x,y))
plt.title('f(x,y)=cos(^2+y^2)')
plt.show()
```

```python
category=['product A','product B','product C','product D','product E']
values=[200, 150, 250, 175, 225]
colors=['red','blue','green','yellow','cyan']
plt.bar(category,values,color=colors)
plt.title('Sales of 5 products')
plt.xlabel('Products')
plt.ylabel('Sales')
plt.show()
```

```python
category=['Category A','Category B','Category C']
T1=[300,200,400]
T2=[750,800,950]
T3=[1100,1300,1500]
T4=[1500,1700,2000]
plt.bar(category,T4,color='yellow',label='T4')
plt.bar(category,T3,color='green',label='T3')
plt.bar(category,T2,color='blue',label='T2')
plt.bar(category,T1,color='red',label='T1')



plt.title('Contribution of 3 categories in 4 time frames')
plt.xlabel('Categories')
plt.legend()
plt.show()
```
